```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import sample
from sklearn.cluster import KMeans
from sklearn.cluster import DBSCAN
%matplotlib inline

pokemon = pd.read_csv('Pokemon.csv')
cluster = pokemon[['Attack','Defense']]

kmeans = KMeans(n_clusters = 2, random_state = 0)
kmeans.fit(cluster)
plt.subplot(2, 2, 1)
plt.title("K-means")
plt.scatter(x = pokemon['Attack'], s = 5, y = pokemon['Defense'], c = kmeans.labels_)

avarage = (sum(pokemon['Attack']) / len(pokemon['Attack']) + sum(pokemon['Defense']) / len(pokemon['Defense'])) / 2
sample = int(avarage / 5)

dbscan = DBSCAN(eps = sample, min_samples = sample, algorithm = 'brute', p = 1)
dbscan.fit(cluster)
plt.subplot(2, 2, 2)
plt.title("DBSCAN")
plt.scatter(x = pokemon['Attack'], s = 5, y = pokemon['Defense'], c = dbscan.labels_)
plt.show()
```

##### I used data of Attack and Defense of the Pokemons for clustering. It's obvious that legendary pokemons are stronger than common.
